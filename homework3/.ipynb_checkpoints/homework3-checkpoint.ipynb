## (1) Databases and sql 
The file "pageviews" contains a table with 4 columns.
1. Load the data file into a table in the empty database "empty_db.sqlite".
2. Write and execute a sql query that computes the sum of the third column of the newly created table.

```python
# write solution here
import sqlite3
import pandas as pd
import csv

con = sqlite3.connect('empty_db.sqlite')
#Drop the table in case it was run earlier and tbl already exist.
con.execute("DROP TABLE tbl")
con.execute("CREATE TABLE tbl(col1 TEXT, col2 TEXT, col3 INTEGER, col4 INTEGER);")

with open('pageviews',encoding="utf8" ) as csvfile:
    csvReader = csv.reader(csvfile, delimiter=' ')
    for rec in csvReader:
        con.execute("INSERT INTO tbl VALUES %s" %str(tuple(rec))) 
csvfile.close()

df=pd.read_sql_query("SELECT SUM(col3) AS SumThirdCol FROM tbl;",con)
con.commit()
con.close()

df
# 233058
```

## (3) Chapter 9, exercise 1

```python
# write solution here
from sklearn import svm 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

#(a) 1 + 3X1 - X2 = 0.
data = {'X1':    [0,1,-1,1,-1,2,-2,3,-3,3,3],
        'X2':    [0,1, 1,2, 2,2, 2,1, 1,2,0],
        'class': [True, True,False, True, False, True, False, True, False, True, True] }
df = pd.DataFrame(data)

# find the maximal separating hyperplane
clf = svm.SVC(kernel='linear',C=10**10)
clf.fit(df[['X1','X2']], df['class'])
# print beta (the hyperplane coefficents) and the support vectors
beta1,beta2 = clf.coef_[0]
beta0 = clf.intercept_[0]
M = 1/np.sqrt(beta1**2 + beta2**2)
print('Part a: 1+3X1-X2 = 0 Hyperplane')
print('beta0=%6.2f \nbeta1=%6.2f \nbeta2=%6.2f'%(beta0,beta1,beta2))
print('Margin:', M)
print('support vectors indexes:', clf.support_)
df.ix[clf.support_,:]
# plot hyperplane, margins and support vectors
fig, ax = plt.subplots(1,2, figsize=(8,4))
ax[0].scatter(df['X1'], df['X2'], c=df['class'], cmap=plt.cm.Paired)
X1_range = [-3,3]
X2_hyperplane = list(map(lambda x1:(-beta0-beta1*x1)/beta2, X1_range))
X2_margin1 = list(map(lambda x1:(-beta0-M*np.sqrt(beta1**2+beta2**2)-beta1*x1)/beta2, X1_range))
X2_margin2 = list(map(lambda x1:(-beta0+M*np.sqrt(beta1**2+beta2**2)-beta1*x1)/beta2, X1_range))

ax[0].plot(X1_range,X2_hyperplane,'-k')
ax[0].plot(X1_range,X2_margin1,'--k')
ax[0].plot(X1_range,X2_margin2,'--k')
ax[0].scatter(clf.support_vectors_[:,0], clf.support_vectors_[:,1], s=80, facecolors='none')
ax[0].set_xlabel('X1')
ax[0].set_ylabel('X2')
ax[0].text(-3, 1, '1+3X1-X2 < 0')
ax[0].text(1, -1, '1+3X1-X2 > 0')
ax[0].set_xlim([-4,4])
ax[0].set_ylim([-4,4]) 
ax[0].set_title('Part a.')


#b. -2+X1+2X2 = 0
data1 = {'X1':   [1,-1,0,-2,-1,-2,2,3,1,1],
        'X2':    [1,-1,0,-1,-2,-2,2,1,3,-3],
        'class': [True,False,False,False,False,False,True,True,True,False] }
df1 = pd.DataFrame(data1)

# find the maximal separating hyperplane
clf1 = svm.SVC(kernel='linear',C=10**10)
clf1.fit(df1[['X1','X2']], df1['class'])
# print beta (the hyperplane coefficents) and the support vectors
beta1_1,beta2_1 = clf1.coef_[0]
beta0_1 = clf1.intercept_[0]
M1 = 1/np.sqrt(beta1_1**2 + beta2_1**2)
print('')
print('Part b: -2+X1+2X2 = 0 Hyperplane')
print('beta0_1=%6.2f \nbeta1_1=%6.2f \nbeta2_1=%6.2f'%(beta0_1,beta1_1,beta2_1))
print('Margin:', M1)
print('support vectors indexes:', clf1.support_)
df1.ix[clf1.support_,:]
# plot hyperplane, margins and support vectors
ax[1].scatter(df1['X1'], df1['X2'], c=df1['class'], cmap=plt.cm.Paired)
X2_hyperplane1 = list(map(lambda x1:(-beta0_1-beta1_1*x1)/beta2_1, X1_range))
X2_margin11 = list(map(lambda x1:(-beta0_1-M1*np.sqrt(beta1_1**2+beta2_1**2)-beta1_1*x1)/beta2_1, X1_range))
X2_margin21 = list(map(lambda x1:(-beta0_1+M1*np.sqrt(beta1_1**2+beta2_1**2)-beta1_1*x1)/beta2_1, X1_range))
ax[1].plot(X1_range,X2_hyperplane1,'-k')
ax[1].plot(X1_range,X2_margin11,'--k')
ax[1].plot(X1_range,X2_margin21,'--k')
ax[1].scatter(clf1.support_vectors_[:,0], clf1.support_vectors_[:,1], s=80, facecolors='none')
ax[1].set_xlabel('X1')
ax[1].set_ylabel('X2')
ax[1].text(-3, -2.5, '-2+X1+2X2 < 0')
ax[1].text(1, 4, '-2+X1+2X2 > 0')
ax[1].set_title('Part b.')

```

```python
# Exercise 9.1 Putting Both Hyperplanes with support vectors on a same graph.
fig, ax = plt.subplots()
ax.scatter(df['X1'], df['X2'], c=df['class'], cmap=plt.cm.Paired)
ax.scatter(df1['X1'], df1['X2'], c=df1['class'], cmap=plt.cm.Paired)
ax.set_xlim([-4,4])
ax.set_ylim([-4,4]) 
ax.set_xlabel('X1')
ax.set_ylabel('X2')
X1_range = [-3,3]
X2_hyperplane = list(map(lambda x1:(-beta0-beta1*x1)/beta2, X1_range))
X2_margin1 = list(map(lambda x1:(-beta0-M*np.sqrt(beta1**2+beta2**2)-beta1*x1)/beta2, X1_range))
X2_margin2 = list(map(lambda x1:(-beta0+M*np.sqrt(beta1**2+beta2**2)-beta1*x1)/beta2, X1_range))
X2_hyperplane1 = list(map(lambda x1:(-beta0_1-beta1_1*x1)/beta2_1, X1_range))
X2_margin11 = list(map(lambda x1:(-beta0_1-M1*np.sqrt(beta1_1**2+beta2_1**2)-beta1_1*x1)/beta2_1, X1_range))
X2_margin21 = list(map(lambda x1:(-beta0_1+M1*np.sqrt(beta1_1**2+beta2_1**2)-beta1_1*x1)/beta2_1, X1_range))
ax.plot(X1_range,X2_hyperplane,'-k',X1_range,X2_hyperplane1,'-r')
ax.plot(X1_range,X2_margin1,'--k',X1_range,X2_margin11,'--r')
ax.plot(X1_range,X2_margin2,'--k',X1_range,X2_margin21,'--r')
ax.scatter(clf.support_vectors_[:,0], clf.support_vectors_[:,1], s=80, facecolors='none')
ax.scatter(clf1.support_vectors_[:,0], clf1.support_vectors_[:,1], s=80, facecolors='none')
```

## (4) Chapter 9, exercise 3

```python
# write solution here
from sklearn import svm 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

#a. Sketch the observations!
data = {'X1':    [3,2,4,1,2,4,4],
        'X2':    [4,2,4,4,1,3,1],
        'class': ['Red', 'Red', 'Red', 'Red', 'Blue', 'Blue', 'Blue'] }
df = pd.DataFrame(data)
fig, ax = plt.subplots()
ax.scatter(df['X1'], df['X2'], c=df['class'], cmap=plt.cm.Paired)
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_title('Exercise 9.3 part a: Sketch Observations')

```

```python
# Exercis 9.3 (b) Optimal separaring hyperplane 

# find the maximal separating hyperplane
clf = svm.SVC(kernel='linear',C=10**10)
clf.fit(df[['X1','X2']], df['class'])
# print beta (the hyperplane coefficents) and the support vectors
beta1,beta2 = clf.coef_[0]
beta0 = clf.intercept_[0]
M = 1/np.sqrt(beta1**2 + beta2**2)
print('beta0=%6.2f \nbeta1=%6.2f \nbeta2=%6.2f'%(beta0,beta1,beta2))
print('Margin:', M)
print('support vectors indexes:', clf.support_)
df.ix[clf.support_,:]

fig, ax = plt.subplots()
ax.scatter(df['X1'], df['X2'], c=df['class'], cmap=plt.cm.Paired)
X1_range = [0.5,4.5]
X2_hyperplane = list(map(lambda x1:(-beta0-beta1*x1)/beta2, X1_range))
ax.plot(X1_range,X2_hyperplane,'-k')
ax.scatter(clf.support_vectors_[:,0], clf.support_vectors_[:,1], s=80, facecolors='none')
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.text(1, 4, 'Red')
ax.text(4, 1, 'Blue')
ax.set_xlim([0.5,4.5])
ax.set_ylim([0.5,4.5]) 
ax.set_title('Exercise 9.3 part b:')

#######################################################
#c. Classification Rule
# 1 - 2X1 + 2X2 > 0 Then 'Red' AND 1-2X1+2X2 < 0 Then 'Blue'
#######################################################
 
```

```python
# Exercise 9.3 (d), (e), (f)

#Margin on maximal margin hyperplane.
fig, ax = plt.subplots()
ax.scatter(df['X1'], df['X2'], c=df['class'], cmap=plt.cm.Paired)
X1_range = [0.5,4.5]
X2_hyperplane = list(map(lambda x1:(-beta0-beta1*x1)/beta2, X1_range))
X2_margin1 = list(map(lambda x1:(-beta0-M*np.sqrt(beta1**2+beta2**2)-beta1*x1)/beta2, X1_range))
X2_margin2 = list(map(lambda x1:(-beta0+M*np.sqrt(beta1**2+beta2**2)-beta1*x1)/beta2, X1_range))
ax.plot(X1_range,X2_hyperplane,'-k')
ax.plot(X1_range,X2_margin1,'--k')
ax.plot(X1_range,X2_margin2,'--k')
ax.scatter(clf.support_vectors_[:,0], clf.support_vectors_[:,1], s=80, facecolors='none')
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.text(1, 4, 'Red')
ax.text(4, 1, 'Blue')
ax.set_xlim([0.5,4.5])
ax.set_ylim([0.5,4.5]) 
ax.set_title('Exercise 9.3 part (d) and (e):')

#f. Slight movement of the seventh observation woulld not affect the maximal margin hyperplane:
#   as that observation (4,1) is outside the margin and support vectors.
```
